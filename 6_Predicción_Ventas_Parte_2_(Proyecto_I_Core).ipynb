**5- Filtrado de Datos**
* Filtra el DataFrame para mostrar solo las filas donde las ventas (Ventas) sean mayores a 50.
* Filtra el DataFrame para mostrar solo las filas donde el precio (Precio) sea menor a 0.5.
* Utilizando el método query(), filtra el DataFrame para mostrar las filas donde el producto sea Manzanas y las ventas sean mayores a 30.

```python
print('\nFiltrado de datos mostrar solo las Ventas y que sea mayor a 50')
df[df['Ventas'] > 50]
```

```python
print('\nFiltrado de datos mostrar solo las Ingreso y que sea menor a 8000')
df[df['Ingreso'] < 8000]
```

```python
print('\nUtilizando el método query(), filtra las filas donde el producto sea 31 y las ventas sean mayores a 30.')
df.query('Productos == 31 & Ventas > 30')
```

**5- Slicing de Datos**
* Selecciona y muestra solo las columnas Producto y Ventas del DataFrame.
* Utilizando loc[], selecciona y muestra las filas de la 5 a la 10 (inclusive) y las columnas Producto y Tienda.
* Utilizando iloc[], selecciona y muestra las primeras 5 filas y las primeras 3 columnas del DataFrame.

```python
print('\nSeleccionar y mostrar solo las columnas Producto y Ventas del DataFrame')
df[['Productos','Ventas']]
```

```python
print('\nUtiliza loc[] para mostrar las filas de la 5 a la 10 (inclusive) y las columnas Producto y Tienda')
df.loc[5:10,['Productos','Tienda']]
```

```python
print('\nUtiliza iloc[] para mostrar las primeras 5 filas y las primeras 3 columnas del DataFrame')
df.iloc[:5,:3]
```
